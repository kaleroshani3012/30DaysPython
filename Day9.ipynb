    - Extend Car into an ElectricCar subclass with battery capacity

```python
class car():

    def __init__(self, brand : str, model : str, year : str, color : str,
                      fuel_type : str = "petrol", mileage : float = 0.0 ):

                      self.brand = brand
                      self.model = model
                      self.year = year
                      self.color = color
                      self.fuel_type = fuel_type
                      self.mileage = mileage

                      self.engine_started = False

    def Engine_Start(self) -> None:

        if not self.engine_started:
            self.engine_started = True
            print(f"{self.brand} {self.model}'s Engine Started")
        else:
            print("Engine Already Started")

    def Engine_Stop(self) -> None:

        if self.engine_started:
            self.engine_started = False
            print(f"{self.brand} {self.model}'s Engine Stopped")
        else:
            print("Engine Already Stopped")

    def Drive(self, distance : float) -> None:

        if self.engine_started:
            self.mileage+=distance
            print(f"You Drove {distance} km. \nTotal mileage is {self.mileage}")
        else:
            print("Please Start the Engine Before Driving")
    
    def display_info(self) -> None:

        print("******Car Information******")
        print(f"Brand       : {self.brand}")
        print(f"Model       : {self.model}")
        print(f"Year        : {self.year}")
        print(f"color       : {self.color}")
        print(f"Fuel Type   : {self.fuel_type}")
        print(f"Mileage     : {self.mileage} km")

        print(f"Engine on?  : {'Yes' if self.engine_started else "No"} ")

```

```python
#subclass inherit from class car

class ElectricCar(car):
    def __init__(self, brand: str, model: str, year: int, color: str, battery_capacity: int, mileage: float = 0.0):
        # Use super() to inherit attributes from Car class
        super().__init__(brand, model, year, color, fuel_type="Electric", mileage=mileage)
        self.battery_capacity = battery_capacity  # new attribute unique to ElectricCar

    # Overriding the display_info method (Polymorphism)
    def display_info(self):
        print("Electric Car Information")
        print(f"Brand            : {self.brand}")
        print(f"Model            : {self.model}")
        print(f"Year             : {self.year}")
        print(f"Color            : {self.color}")
        print(f"Battery Capacity : {self.battery_capacity} kWh")
        print(f"Mileage          : {self.mileage} km")
        print(f"Engine On?       : {'Yes' if self.engine_started else 'No'}")


car1 = car("Hyundai", "i20", 2021, "Blue")
car1.Engine_Start()
car1.Drive(60)
car1.display_info()
car1.Engine_Stop()

print("\n")

# Electric car instance
e_car = ElectricCar("Tesla", "Model 3", 2023, "White", battery_capacity=75)
e_car.Engine_Start()
e_car.Drive(120)
e_car.display_info()
e_car.Engine_Stop()
```
