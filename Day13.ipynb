```python
def is_balanced_paranthesis(expression, show_stack = True, custom_pairs=None):

    stack=[]
    default_pairs = { '(' : ')' , '[' : ']' , '{' : '}' , '<' : '>'}
    pairs = custom_pairs if custom_pairs else default_pairs
    reverse_pairs = {v: k for k , v in pairs.items()}


    for id , char in enumerate(expression):

        if char in pairs:
            stack.append(char)
            
            if show_stack:
                print(f"{id} : Push '{char}' -> Stack : {stack} ")

        elif char in reverse_pairs:

            if not stack:
                print(f" Error, closing '{char}' Found but stack has not opening character.")
                return False
            
            top = stack[-1]
            if reverse_pairs[char] == top:
                stack.pop()

                if show_stack:
                    print(f"{id} : Pop {top} for closing '{char}' -> Stack : {stack}")

            else:
                print(f"{id} : Error: Mismatched! Expected closing for {top} but got '{char}' ")
                return False
            
    if stack:
        print(f" Error -> Stack not empty at end -> Remaining {stack}")
        return False
    
    print("Yay,Expression is Balanced.")
    return True
```

```python
def expression_check():

    print("****** Balanced Bracket Check ******")
    print("Supports (), [], {}, <>")
    print("Type 'Exit' to Quit\n")

    valid_brackets = set("()[]{}<>")

    while True:

        expression = input("Enter Expression: ").strip()

        if expression.lower() == 'exit':
            print("*******Bye,Exiting Checker*******")
            break

        if not expression:
            print("Empty Input, Try Again!!!")
            continue
        if not any (ch in valid_brackets for ch in expression):
            print("oops, No Bracket Found.\n Please Enter an Expression with Brackets.\n")
            continue
        
        print(f"Wait, Checking Expression -> {expression} ")
        is_balanced_paranthesis(expression)
        print('-'*50)

if __name__ == "__main__":
    expression_check()
```
